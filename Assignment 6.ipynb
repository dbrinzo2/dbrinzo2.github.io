```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets
import bqplot
import pandas as pd
```

```python
buildings = pd.read_csv('https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv',
               na_values={'Square Footage': 0, 
                         'Year Acquired': 0,
                         'Year Constructed': 0, 
                         'Floors':0})
```

```python
buildings
```

```python
buildings_index = pd.pivot_table(buildings, index = ['Agency Name'], values = ['Square Footage'], columns = ["County"], aggfunc = np.mean)
buildings_index
```

```python
county = buildings_index.columns.levels[1].to_list()
#Colors
col_sc = bqplot.ColorScale(scheme = "Reds")
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

#Heat Map
heat_map = bqplot.GridHeatMap(color = np.log10(buildings_index.values), row = buildings_index.index, column = county, 
                             scales = {'color' : col_sc, 'row' : y_sc, 'column' : x_sc},
                             interactions = {'click':'select'},
                             anchor_style = {'fill':'blue'},
                             selected_style = {'opacity' : 1.0},
                             unselected_style = {'opacity' : 1.0})
                            
#Axis
c_ax = bqplot.ColorAxis(scale = col_sc, orientation = 'vertical', side = 'right')
x_ax = bqplot.Axis(scale = x_sc, label = 'County', tick_rotate = 270, 
                   tick_style = {'font-size' : '8px', "tick_offset" : '100px', "text_anchor" : "center"})
y_ax = bqplot.Axis(scale = y_sc, orientation = 'vertical', tick_style = {'font-size' : '8px'})       

#Figure object
fig = bqplot.Figure(marks=[heat_map], axes=[c_ax, x_ax, y_ax], figsize = (20, 12))
fig

#Interact
mySelectedLabel = ipywidgets.Label()
mySelectedLabel.value = "Mean Point"
def on_selected(change):
    i,j = change['owner'].selected[0]
    mask = (buildings['Agency Name'] == buildings_index.index[i]) & \
            (buildings['County'] == buildings_index.columns.levels[1][j])
    buildings_subset = buildings[mask]
    grouped = buildings_subset.groupby("Year Acquired")["Square Footage"].sum()
    years = grouped.index
    sqft = grouped.values
    bar.x = years
    bar.y = sqft

#Linking
heat_map.observe(on_selected, 'selected') 
```

```python
# Barplot
x_scl = bqplot.DateScale()
y_scl = bqplot.LinearScale()
ax_xcl = bqplot.Axis(label = 'Year', scale=x_scl, tick_style = {'font-size' : '8px'})
ax_ycl = bqplot.Axis(label = 'Square Footage in Year Acquired', scale = y_scl, 
                    orientation = 'vertical', side = 'left', tick_style = {'font-size' : '8px'})

# start with defaults
i,j = 0,0
mask = (buildings['Agency Name'] == buildings_index.index[i]) & \
       (buildings['County'] == buildings_index.columns.levels[1][j])

# new, subset dataframe
buildings_subset = buildings[mask]

# grab years and sqft from pandas
grouped = buildings_subset.groupby("Year Acquired")["Square Footage"].sum()
years = grouped.index
sqft = grouped.values

# plot
bar = bqplot.Bars(x = years, y = sqft,
                    scales = {'x': x_scl, 'y': y_scl})

fig_bar = bqplot.Figure(marks = [bar], axes = [ax_ycl, ax_xcl])
fig_bar
```

```python
fig.layout.min_width='600px'
fig_bar.layout.min_width = '500px'
figures = ipywidgets.HBox([fig, fig_bar])
myDash = ipywidgets.VBox((mySelectedLabel, figures))
myDash
```
